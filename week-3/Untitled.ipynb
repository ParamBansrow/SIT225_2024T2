```python
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting in Jupyter
%matplotlib inline  

# Load the CSV file
df = pd.read_csv("Untitled-distance_cm.csv")  # Replace with your actual CSV file name

# Display column names to verify
print("Column Names:", df.columns)

# Rename columns for consistency (optional)
df.rename(columns={"time": "timestamp", "value": "distance"}, inplace=True)

# Convert timestamp column to datetime format
df["timestamp"] = pd.to_datetime(df["timestamp"], errors='coerce')

# Remove rows with missing values
df = df.dropna()

# Remove negative or zero distance values (if necessary)
df = df[df["distance"] > 0]

# Plot the graph
plt.figure(figsize=(12, 6))
plt.plot(df["timestamp"], df["distance"], marker="o", linestyle="-", color="b", label="Distance (cm)")

# Customize the graph
plt.xlabel("Time")
plt.ylabel("Distance (cm)")
plt.title("Ultrasonic Sensor Distance Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Save the graph as an image
plt.savefig("ultrasonic_plot.png", dpi=300, bbox_inches='tight')

# Show the graph
plt.show()
```
