```python
import pandas as pd
import matplotlib.pyplot as plt


filename = "dht22_data.csv"  
df = pd.read_csv(filename)


df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%Y%m%d%H%M%S")


plt.figure(figsize=(12, 6))
plt.plot(df["Timestamp"], df["Temperature (C)"], label="Temperature (Â°C)", color="red", marker='o', linestyle='-')
plt.plot(df["Timestamp"], df["Humidity (%)"], label="Humidity (%)", color="blue", marker='s', linestyle='-')


plt.xlabel("Time")
plt.ylabel("Value")
plt.title("DHT22 Temperature & Humidity Over Time")
plt.legend()
plt.xticks(rotation=45)  
plt.grid()


plt.show()
```
